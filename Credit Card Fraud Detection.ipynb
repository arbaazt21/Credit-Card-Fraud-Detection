# Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

# Loading Dataset

```python
df = pd.read_csv("creditcard.csv")
```

```python
df.head()
```

```python
# Dataset Information

df.info()
```

```python
# Checking the number of missing values in each column

df.isna().sum()
```

```python
# Distribution of Legit-Transaction and Fradulent-Transaction

df["Class"].value_counts()
```

### The Dataset is highly Imbalanced 

```python
# Separating the Data for Analysis

legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
legit.shape
```

```python
fraud.shape
```

```python
# Statistical Measures about the Data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both Transactions

df.groupby("Class").mean()
```

# Under-Sampling

```python
legit_sample = legit.sample(492)
```

# Concatenating two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset["Class"].value_counts()
```

```python
new_dataset.groupby("Class").mean()
```

# Extracting Features and Target Columns

```python
# In the dataset,

# Features = include all columns, except the Class column (So for this, we need to drop the class column).
# Target = Class

# x = Features 
# y = Target
```

```python
x = new_dataset.drop(columns="Class", axis=1)
y = new_dataset["Class"]
```

```python
x
```

```python
y
```

# Split the Dataset into Train and Test

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

# Model Training

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
# Training the Logistic Regression Model with Training Data

model.fit(x_train, y_train)
```

# Model Evaluation

#### Accuracy Score

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, f1_score
```

```python
# Accuracy on Training Data

x_train_prediction = model.predict(x_train)
```

```python
training_data_accuracy = accuracy_score(x_train_prediction, y_train)

print("Accuracy Score of Training Data is: ", training_data_accuracy)
```

```python
# Accuracy on Test Data

x_test_prediction = model.predict(x_test)
```

```python
test_data_prediction = accuracy_score(x_test_prediction, y_test)

print("Accuracy Score of Test Data is: ",test_data_prediction)
```

#### Confusion Matrix

```python
conf_matrix = confusion_matrix(x_test_prediction, y_test)

print("Confusion Matrix:\n", conf_matrix)
```

#### Calculate Precision, Recall, and F1-Score

```python
precision = precision_score(x_test_prediction, y_test)

recall = recall_score(x_test_prediction, y_test)

f1 = f1_score(x_test_prediction, y_test)
```

```python
print("Precision: {:.4f}".format(precision))
print("Recall: {:.4f}".format(recall))
print("F1-Score: {:.4f}".format(f1))
```
